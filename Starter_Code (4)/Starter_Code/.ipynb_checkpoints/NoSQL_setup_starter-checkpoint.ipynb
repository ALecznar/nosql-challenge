5. Some of the number values are stored as strings, when they should be stored as numbers.

Use `update_many` to convert `latitude` and `longitude` to decimal numbers.

```python
# Change the data type from String to Decimal for longitude and latitude
for document in establishments.find(query):
    longitude_decimal = Decimal(document['geocode']['longitude'])
    latitude_decimal = Decimal(document['geocode']['latitude'])
    
    update_query = {'_id': document['_id']}
    update_values = {
        '$set': {
            'geocode.longitude': longitude_decimal,
            'geocode.latitude': latitude_decimal
        }
    }
    establishments.update_one(update_query, update_values)
```

Use `update_many` to convert `RatingValue` to integer numbers.

```python
query = {'RatingValue': {'$type': 'string', '$regex': '^[0-9]+$'}}
update_operation = [
    {
        '$set': {
            'RatingValue': {'$toInt': '$RatingValue'}
        }
    }
]

result = establishments.update_many(query, update_operation)
```

```python
# Set non 1-5 Rating Values to Null
non_ratings = ["AwaitingInspection", "Awaiting Inspection", "AwaitingPublication", "Pass", "Exempt"]
establishments.update_many({"RatingValue": {"$in": non_ratings}}, [ {'$set':{ "RatingValue" : None}} ])
```

```python
# Check that the coordinates and rating value are now numbers
for document in establishments.find():
    longitude_type = type(document['geocode']['longitude'])
    latitude_type = type(document['geocode']['latitude'])
    rating_value_type = type(document['RatingValue'])

    print(f"Document ID: {document['_id']}")
    print(f"Longitude data type: {longitude_type}")
    print(f"Latitude data type: {latitude_type}")
    print(f"RatingValue data type: {rating_value_type}")
```
